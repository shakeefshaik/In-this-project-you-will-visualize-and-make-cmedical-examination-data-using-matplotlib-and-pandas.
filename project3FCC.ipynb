Sure! Here's a brief explanation of the columns in the medical dataset and some general medical information:

1. **Age**: The age of the patient in days.
2. **Height**: The height of the patient in centimeters.
3. **Weight**: The weight of the patient in kilograms.
4. **Gender**: The gender of the patient (categorical code).
5. **Systolic Blood Pressure (ap_hi)**: The pressure in your arteries when your heart beats. Normal range is typically around 90-120 mmHg.
6. **Diastolic Blood Pressure (ap_lo)**: The pressure in your arteries when your heart rests between beats. Normal range is typically around 60-80 mmHg.
7. **Cholesterol**: 
   - 1: Normal
   - 2: Above normal
   - 3: Well above normal
8. **Glucose (gluc)**: 
   - 1: Normal
   - 2: Above normal
   - 3: Well above normal
9. **Smoking (smoke)**: Indicates if the patient smokes (binary).
10. **Alcohol Intake (alco)**: Indicates if the patient consumes alcohol (binary).
11. **Physical Activity (active)**: Indicates if the patient is physically active (binary).
12. **Cardiovascular Disease (cardio)**: Indicates the presence (1) or absence (0) of cardiovascular disease.

### General Medical Information:
- **BMI (Body Mass Index)**: Calculated as weight (kg) / (height (m))^2. A BMI over 25 is considered overweight.
- **Cholesterol Levels**: 
  - Normal: Less than 200 mg/dL
  - Borderline high: 200-239 mg/dL
  - High: 240 mg/dL and above
- **Glucose Levels**: 
  - Normal fasting blood sugar: 70-99 mg/dL
  - Prediabetes: 100-125 mg/dL
  - Diabetes: 126 mg/dL and above


```python
df=pd.read_csv(r"C:\Users\sunny\Downloads\medical_examination.csv")
df.head()
```

Adding a column i.e OverWeight

```python

# Convert height from cm to meters
df['height_m'] = df['height'] / 100

# Calculate BMI
df['BMI'] = df['weight'] / (df['height_m'] ** 2)

# Determine if the person is overweight (BMI > 25)
df['overweight'] = (df['BMI'] > 25).astype(int)

# Display the first few rows of the DataFrame to verify
print(df.head())

```

```python
df
```

Normalize cholesterol column

```python

# Normalize cholesterol column
df['cholesterol'] = df['cholesterol'].apply(lambda x: 0 if x == 1 else 1)

# Normalize gluc column
df['gluc'] = df['gluc'].apply(lambda x: 0 if x == 1 else 1)

# Display the first few rows of the DataFrame to verify
print(df.head())
```

Create DataFrame for cat plot using pd.melt

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create DataFrame for cat plot using pd.melt
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active', 'overweight'])

# Group and reformat the data to split it by cardio and show the counts of each feature
df_cat = df_cat.groupby(['cardio', 'variable', 'value']).size().reset_index(name='total')

# Draw the categorical plot
fig = sns.catplot(x='variable', y='total', hue='value', col='cardio', data=df_cat, kind='bar', height=5, aspect=1)
fig.set_axis_labels('Variable', 'Total')
fig.set_titles('Cardio {col_name}')
fig.despine(left=True)
plt.show()
```

Clean the data and Calculate the correlation matrix

```python
# Clean the data
df_heat = df[(df['ap_lo'] <= df['ap_hi']) &
             (df['height'] >= df['height'].quantile(0.025)) &
             (df['height'] <= df['height'].quantile(0.975)) &
             (df['weight'] >= df['weight'].quantile(0.025)) &
             (df['weight'] <= df['weight'].quantile(0.975))]

# Calculate the correlation matrix
corr = df_heat.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(12, 10))

# Draw the heat map
sns.heatmap(corr, mask=mask, annot=True, fmt=".1f", cmap='coolwarm', ax=ax)

# Show the plot
plt.show()
```
